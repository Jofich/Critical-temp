```python
df = {}
coun = []
for critical_temp in range(1,4):
    for i in formula_train[formula_train['critical_temp'] < critical_temp]['material']:
        for b in i:
            if b.isdigit():
                i = i.replace(b,"")
        i = i.split(".")
        if i[-1] == '':
            i.pop(-1)
        for element in (i):
            for index in range(len(element)-1):
                if element[index].istitle() and not element[index + 1].istitle():
                    name = element[index] + element[index + 1]
                if element[index].istitle() and element[index + 1].istitle():
                    name = element[index]
                if name not in df:
                    df[name] = 1
                else:
                    df[name] += 1
    df = dict(sorted(df.items(), key=lambda item: -item[1]))
    a = {}
    for ind in range(4):
        a[list(df.keys())[ind]] = df[list(df.keys())[ind]]
    coun.append([a,critical_temp])
       
df = dict(sorted(df.items(), key=lambda item: -item[1])) 
popular = list(coun[0][0].keys())
```

```python
df_corr = train.corr()

f, ax = plt.subplots(figsize=(15, 10))
sns.heatmap(df_corr, cmap = "BrBG",ax=ax)
plt.show()
```

```python
correlated_features = set()

for i in range(len(df_corr.columns)):
    for j in range(i):
        if abs(df_corr.iloc[i, j]) < 0.01:
            colname = df_corr.columns[i]
            correlated_features.add(colname)
            
if "critical_temp" in correlated_features:
    train_corr_best = train[correlated_features]
    correlated_features.remove("critical_temp")
    test_corr_best = test[correlated_features]
else:
    test_corr_best = test[correlated_features]
    correlated_features.add("critical_temp")
    train_corr_best = train[correlated_features]


```

```python
test_corr_best
```

```python
X = train_corr_best.drop(columns = 'critical_temp')
y = train_corr_best['critical_temp']
```

```python
def solution(array,name):
    df = pd.DataFrame(data = {"critical_temp":array}).reset_index()
    df.to_csv(f'{name}.csv',index = False)
```

```python
from sklearn.linear_model import LinearRegression

LR = LinearRegression()

LR.fit(X, y)
```

```python
y_pred = LR.predict(test_corr_best)
```

```python
df = pd.DataFrame({"critical_temp":y_pred})
```

```python
df.to_csv('LR_1.csv')
```
