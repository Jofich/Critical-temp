```python
df = {}
coun = []
for critical_temp in range(1,4):
    for substance in formula_train[formula_train['critical_temp'] < critical_temp]['material']:
        for b in substance:
            #удаление индексов
            if b.isdigit():
                substance = substance.replace(b,"")
        
        substance = substance.split(".")
        if substance[-1] == '':
            substance.pop(-1)
        for element in (substance):
            for index in range(len(element)-1):
                if element[index].istitle() and not element[index + 1].istitle():
                    name = element[index] + element[index + 1]
                if element[index].istitle() and element[index + 1].istitle():
                    name = element[index]
                if name not in df:
                    df[name] = 1
                else:
                    df[name] += 1
    df = dict(sorted(df.items(), key=lambda item: -item[1]))
    a = {}
    for ind in range(4):
        a[list(df.keys())[ind]] = df[list(df.keys())[ind]]
    coun.append([a,critical_temp])
       
df = dict(sorted(df.items(), key=lambda item: -item[1])) 
popular = list(coun[0][0].keys())
popular
```
